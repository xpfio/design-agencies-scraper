# Simple Crawler studio-index

```python
from time import sleep
from random import randint
from selenium import webdriver
from pyvirtualdisplay import Display
```

```python
url_to_crawl = "http://studio-index.co/"
```

```python
print('starting driver...')
display = Display(visible=1, size=(1280, 720))
display.start()
driver = webdriver.Chrome()
sleep(4)
```

```python
print('getting page...')
driver.get(url_to_crawl)
sleep(randint(2,3))
```

```python
driver.set_window_size(1280, 720)
print driver.get_window_size()
```

```python
print('grabbing list of items...')
list_studio = []
print len(driver.find_elements_by_class_name("region"))
data = []

for region in driver.find_elements_by_class_name("region"):
    print len(region.find_elements_by_class_name("ng-binding"))
    for div in region.find_elements_by_class_name("ng-binding"):
        if div.tag_name != 'h2':
            #simple way to get the table
            data.append(div.text)
            if len(data) >= 4:
                list_studio.append({
                    'country':data[0],
                    'city':data[1],
                    'name':data[2],
                    'url':data[3]
                })
                data = []
```

```python
import pandas as pd
print len(list_studio)
df = pd.DataFrame.from_dict(list_studio)
df
```

```python
import datetime
# Export the Data
df.to_csv('data '+ str(datetime.datetime.now()) + '.csv', header=True, encoding='utf-8')
```

```python
print('closing driver...')
display.stop()
driver.quit()
print('closed!')
```
